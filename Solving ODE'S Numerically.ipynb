# 1. 
Solve the IVP

$$\frac{dy}{dx}=2y(1-x)$$ ,$$x \in [0,10],y_0=1$$

using all the three methods (1) Taylor method of order 2, (2) Euler’s method, and (3)
RK2 method, with a step size of h = 0.02.

(1) Taylor method of order 2

```python
def f(x,y): # the function f(x,y)
    return 2*y*(1-x)

def f_x(x,y): # the partial derivative of f(x,y) w.r.t. x
    return -2*y

def f_y(x,y): # the partial derivative of f(x,y) w.r.t. y
    return 2*(1-x)

def y_i(x,y,h): # gives y_i+1 using Taylor Method
    return y+h*f(x,y)+((h**2)*(0.5))*(f_x(x,y)+f_y(x,y)*f(x,y))

h=0.02  # step size
n=(10-0)/(0.02)

x_0=0
x=[]

for i in range(int(n+1)):
    x.append(x_0 + i*h)

y_t=[1]

for j in range(int(n)):
    r=y_i(x[j],y_t[j],0.02)
    y_t.append(round(r,5))
print("The numerical solution y(x_i) using Taylor method of order 2",y_t )
```

(2) Euler’s method

```python
y_e=[1]
for j in range(int(n)):
    r=y_e[j]+h*f(x[j],y_e[j]) # gives y_i+1 using Euler's Method
    y_e.append(round(r,5))
print("The numerical solution y(x_i) using Euler's method of order 2",y_e )
```

(3) RK2 method

```python
y_rk=[1]
for i in range(int(n)):
    k1=h*f(x[i],y_rk[i])
    k2=h*f(x[i]+h,y_rk[i]+k1)
    r=y_rk[i]+0.5*(k1+k2)  # using RK2 Method
    y_rk.append(round(r,5))
print("The numerical solution y(x_i) using RK2 Method",y_rk )
```

# 2. 
For the ODE given above, obtain the exact analytic solution.

Exact Solution :

$$\frac{dy}{dx}=2y(1-x)$$

$$\int \frac{dy}{y}=\int 2(1-x)dx$$
$$\log{y}=2(x-\frac{x^2}{2})+C$$

Using initial condition, y=1 at x=0 .We get ,C=0.
Therefore,the solution of the given ODE is $ y(x)=e^{(2x-x^2)}$

```python
import numpy as np
def f_exact(x):
    return np.exp(2*x-x**2)
y_exact=[]
for j in range(int(n+1)):
    y_exact.append(f_exact(x[j]))
```

Compare the numerical solutions obtianed using the three methods in the first question
with the exact solution you derived in the second question, by plotting all the four and
also their relative differences w.r.t. the exact solution separately.

Plot of numerical solutions obtianed using the three methods in the first question with the exact solution :

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y_t,"r")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("y(x) using Taylor method of order 2 ")
plt.axis([0,10,0,3])
plt.grid()
plt.show()

plt.plot(x,y_e,"b")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("y(x) using Euler's Method")
plt.axis([0,10,0,3])
plt.grid()
plt.show()

plt.plot(x,y_t,"g")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("y(x) using RK2 method")
plt.axis([0,10,0,3])
plt.grid()
plt.show()

plt.plot(x,y_exact,"black")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("y(x) using exact solution")
plt.axis([0,10,0,3])
plt.grid()
plt.show()
```

Plot of their relative differences w.r.t. the exact solution :

```python
dif_rk=[]
dif_e=[]
dif_t=[]

for i in range(int(n+1)):
    dif_rk.append((abs(round(y_rk[i]-y_exact[i],5))))
    dif_e.append((abs(round(y_e[i]-y_exact[i],5))))
    dif_t.append((abs(round(y_t[i]-y_exact[i],5))))

plt.plot(x,dif_t,"r--")
plt.xlabel("x")
plt.ylabel("relative difference")
plt.title("Relative difference b/w y(x) using Taylor Method & exact solution")
plt.grid()
plt.show()

plt.plot(x,dif_e,"b--")
plt.xlabel("x")
plt.ylabel("relative difference")
plt.title("Relative difference b/w y(x) using Euler Method & exact solution")
plt.grid()
plt.show()

plt.plot(x,dif_rk,"g--")
plt.xlabel("x")
plt.ylabel("relative difference")
plt.title("Relative difference b/w y(x) using RK2 Method & exact solution")
plt.grid()
plt.show()
```
