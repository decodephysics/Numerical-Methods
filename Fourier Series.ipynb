# Fourier Series

$ f(x)$ is a periodic function s.t. $f(x+2\pi)=f(x)$ where $ -\pi \leq x \leq \pi$

$ f(x) = a_0 + \sum_{n=0}^{\infty}\cos{nx} +\sum_{n=0}^{\infty}\sin{nx} $

- $ a_0=\frac{1}{2\pi}\int_{-\pi}^{\pi} f(x) dx$
- $ a_n=\frac{1}{\pi}\int_{-\pi}^{\pi} f(x)\cos {nx} dx$
- $ b_n=\frac{1}{\pi}\int_{-\pi}^{\pi} f(x)\sin{nx} dx$

$f(x)=x $ where $ -\pi \leq x \leq \pi$<br><br>
On solving, We get $ a_0 =0 ,a_n=0 ,b_n=\frac{-2*(-1)^n}{n}$

```python
import numpy as np
x=np.linspace(-1*np.pi,np.pi,num=100)
import matplotlib.pyplot as plt
plt.plot(x,x)
plt.plot([-3,0,3],[0,0,0],"r--")
plt.show()
```

```python
def f(n,x):
    sum=0
    for i in range(1,n+1):
        b_i=(-2*(-1)**i)/i
        sum+=b_i*np.sin(i*x)
    return sum
```

```python
plt.plot(x,f(1,x)) #n=1
plt.show()
```

```python
plt.plot(x,f(2,x)) #n=2
plt.show()
```

```python
plt.plot(x,f(3,x)) # n=3
plt.show()
```

```python
plt.plot(x,f(5,x)) #n=5
plt.show()
```

```python
plt.plot(x,f(10,x))#n=10
plt.show()
```

```python
plt.plot(x,f(100,x))#n=100
plt.show()
```

We observe that as we increase the value of n ,Fourier series converge to f(x).
