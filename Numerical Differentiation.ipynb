```python
import numpy as np
import math

def f(x):  # define the given f(x)
    return np.exp(x)

def round_up(x): # round up the number to four significant digits
    power = -int(math.floor(math.log10(abs(x)))) + 3
    factor = (10 ** power)
    return round(x * factor) / factor
```

```python
x=np.linspace(0,10,num=11)
h=np.array([1,1e-1,1e-2,1e-3,1e-4])
```

```python
def F1(x,h): # gives first order derivative 
    y=(f(x+h)-f(x-h))/(2*h)
    return round_up(y)
```

```python
for i in x:
    for j in h:
        res=F1(i,j)
        print("f'3(x=",i,",h=",j,")=",res)
    print("f'3(x=",i,"exact)=",round_up(f(i)),"\n")
```

```python
def F2(x,h): # gives second order derivative 
    y=(f(x+h)+f(x-h)-2*f(x))/(h**2)
    return round_up(y)
```

```python
for i in x:
    for j in h:
        res=F2(i,j)
        print("f\"3(x=",i,",h=",j,")=",res)
    print("f\"3(x=",i,"exact)=",round_up(f(i)),"\n")
```
